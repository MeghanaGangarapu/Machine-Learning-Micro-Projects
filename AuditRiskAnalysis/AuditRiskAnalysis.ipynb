### Feature Engineering
New interaction variables are added to capture the relation between variables. Interaction terms in regression model can greatly expand understanding of the relationships among the variables in the model and allows more hypotheses to be tested.
For example: The effect of Loss (Amount of loss suffered by the firm in last year) on Audit_Risk is not constant. It is related with the historical loss suffered by the firm in the last 10 years. This relation can be captured using interaction terms. Similarly for other interaction variables.

```python
riskdf['MV_SectorScore'] = riskdf['Money_Value']*riskdf['Sector_score']
riskdf['ParaAB'] = riskdf['PARA_A']*riskdf['PARA_B']
riskdf['ParaA_Number'] = riskdf['PARA_A']*riskdf['numbers']
riskdf['ParaB_Number'] = riskdf['numbers']*riskdf['PARA_B']
riskdf['Total_History'] = riskdf['TOTAL']*riskdf['History']
riskdf['Loss_History'] = riskdf['Loss']*riskdf['History']
riskdf['District_History'] = riskdf['District_Loss']*riskdf['History']
riskdf['Sector_History'] = riskdf['Sector_score']*riskdf['History']
riskdf['Sector_District'] = riskdf['District_Loss']*riskdf['Sector_score']
```

### Ridge Regression

```python
## logistic regression without cross-validation and gridsearch
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
c_range = [0.001, 0.01, 0.1, 1, 10]
train_score_l1 = []
train_score_l2 = []
test_score_l1 = []
test_score_l2 = []

for c in c_range:
    log_l1 = LogisticRegression(penalty = 'l1', C = c)
    log_l2 = LogisticRegression(penalty = 'l2', C = c)
    log_l1.fit(X_train, y_train)
    log_l2.fit(X_train, y_train)
    train_score_l1.append(log_l1.score(X_train, y_train))
    train_score_l2.append(log_l2.score(X_train, y_train))
    test_score_l1.append(log_l1.score(X_test, y_test))
    test_score_l2.append(log_l2.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(c_range, train_score_l1, label = 'Train score, penalty = l1')
plt.plot(c_range, test_score_l1, label = 'Test score, penalty = l1')
plt.plot(c_range, train_score_l2, label = 'Train score, penalty = l2')
plt.plot(c_range, test_score_l2, label = 'Test score, penalty = l2')
plt.legend()
plt.xlabel('Regularization parameter: C')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.grid()
```

```python
print("Accuracy Score train:",train_score_l1[c_range.index(1)])
print("Accuracy Score test:",test_score_l1[c_range.index(1)])
```
